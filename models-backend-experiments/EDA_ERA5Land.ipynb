## Section 1. Load "ERA5 LAND" dataset

### Section 1.1. Setting dataset path

```python
DATASETS_FOLDER = "/home/ecarreno/SSC-Projects/a_DATASETS/ERA5-Land"
subfolder_dataset= "2024_allmonths"
dataset = "data_stream-moda.nc"

dataset_path = os.path.join(
    DATASETS_FOLDER,
    subfolder_dataset,
    dataset,
)

print("This file exist? {}".format(os.path.exists(dataset_path)))
```

### Section 1.2. Load dataset with XARRAY

```python
ds = xr.open_dataset(dataset_path, chunks={"valid_time": 100})
ds
```

```python
print(ds)
```

```python
print(type(ds["t2m"].data))
```

- **Dimensions** specify the number of element of each data coordinate:
   - valid_time (12): one single date for each month.
   - latitute (1801): values between -90 to 90 degrees.
   - longitude (3600): values between 0 and 360 degrees
- **Attributes** provide metadata about the file such as the provenance (ecmf)
- **Variables** contains the actual data. In this file we just have the temperature above two meters (t2m).

```python
ds_celsius = ds.t2m - 273.15
```

```python
ds_avg = ds_celsius.mean(dim="valid_time")
```

```python
print(ds_avg)
```

```python
# Define the map projection (how does the map look like)
ax = plt.axes(projection=ccrs.EqualEarth())
# ax is an empty plot. We now plot the variable sw_avg onto ax
ds_avg.plot(ax=ax, transform=ccrs.PlateCarree()) 
# the keyword "transform" tells the function in which projection the data is stored 
ax.coastlines(); ax.gridlines(); # Add gridlines and coastlines to the plot
```

```python
ds_avg.plot()
```

```python
ds_avg.mean(dim="longitude").plot()
```

```python
ds_avg.median(dim="latitude").plot()
```
